```python
#1.1 Write a Python Program(with class concepts) to find the area of the triangle using the below formula.
#area = (s*(s-a)*(s-b)*(s-c)) ** 0.5
#Function to take the length of the sides of triangle from user should be defined in the parent
#class and function to calculate the area should be defined in subclass.

class triangle:
    def __init__(self,a,b,c):
        self.a = float(a)
        self.b = float(b)
        self.c = float(c)
a = int(input("Enter the length of side1: "))
b = int(input("Enter the length of side2: "))
c = int(input("Enter the length of side3: "))

class area(triangle):
    def __init__(self,a,b,c):
        super().__init__(a,b,c)

    def get_area(self):
        s = (a + b + c) / 2
        return (s*(s-a)*(s-b)*(s-c)) ** 0.5        

t = area(a,b,c)
print("Area of the triangle with the sides {},{},{} is  {}".format(a,b,c,t.get_area()))
```

```python
#1.2 Write a function filter_long_words() that takes a list of words and an integer n and returns
#the list of words that are longer than n.

def filter_long_words(wordlist, integer_n):
    return (word for word in wordlist if len(word) >= integer_n)

def inputs():
    words = input("Enter words, separated by spaces: ").split()
    integer_n = int(input("Minimum length of words to keep: "))
    print("Words longer than {} are {}.".format(integer_n,
          ', '.join(filter_long_words(words, integer_n))))

inputs()
```

```python
#2.1 Write a Python program using function concept that maps list of words into a list of integers
#representing the lengths of the corresponding words.

def find_longest_word(words_list):
    word_len = []
    for n in words_list:
        word_len.append(len(n))
    return word_len
result = find_longest_word(["Dhaya", "Ineuron", "DataScientist"])
print(result)
```

```python
#2.2 Write a Python function which takes a character (i.e. a string of length 1) and returns True if
#it is a vowel, False otherwise.

def vowels(list_1):
    inp = input("Enter a character: ")
    if inp in list_1:
        print("True")
    else:
        print("False")
vowels(["a", "e", "i", "o", "u", "A", "E", "I", "O", "U"])
```
